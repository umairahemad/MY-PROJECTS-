## Squat Counter


```python
import cv2
import mediapipe as mp

cap = cv2.VideoCapture(0)

mp_pose = mp.solutions.pose
counter = 0
stage = None
THRESHOLD = 0.03  # You can adjust this for sensitivity

with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False

        results = pose.process(image)

        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

        mp.solutions.drawing_utils.draw_landmarks(
            image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)

        if results.pose_landmarks:
            hip_y = results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_HIP].y
            knee_y = results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_KNEE].y

            print(f"Hip Y: {hip_y:.3f}, Knee Y: {knee_y:.3f}, Stage: {stage}")

            # Squat down: hip below knee
            if hip_y > knee_y + THRESHOLD:
                stage = "down"
            # Stand up: hip above knee
            if hip_y < knee_y - THRESHOLD and stage == "down":
                stage = "up"
                counter += 1

        cv2.putText(image, f'Squats: {counter}', (10, 30),
                    cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0), 2)
        cv2.imshow("Mediapipe Feed", image)

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import mediapipe as mp
import time

# --- Exercise Logic Functions ---
def squat_counter(results, stage, counter, last_time, cheat_count, THRESHOLD=0.03, min_depth=0.05):
    hip_y = results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_HIP].y
    knee_y = results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_KNEE].y
    now = time.time()
    cheat = False

    # Cheat: not deep enough
    if abs(hip_y - knee_y) < min_depth:
        cheat = True

    # Cheat: rep too fast
    if last_time and (now - last_time < 0.5):
        cheat = True

    if hip_y > knee_y + THRESHOLD:
        stage = "down"
    if hip_y < knee_y - THRESHOLD and stage == "down":
        stage = "up"
        if not cheat:
            counter += 1
            last_time = now
        else:
            cheat_count += 1
            last_time = now
    return stage, counter, cheat_count, last_time, cheat

def bicep_curl_counter(results, stage, counter, last_time, cheat_count, THRESHOLD=0.1, min_range=0.08):
    wrist_y = results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_WRIST].y
    shoulder_y = results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_SHOULDER].y
    now = time.time()
    cheat = False

    # Cheat: not enough range
    if abs(wrist_y - shoulder_y) < min_range:
        cheat = True

    # Cheat: rep too fast
    if last_time and (now - last_time < 0.5):
        cheat = True

    if wrist_y < shoulder_y - THRESHOLD:
        stage = "up"
    if wrist_y > shoulder_y + THRESHOLD and stage == "up":
        stage = "down"
        if not cheat:
            counter += 1
            last_time = now
        else:
            cheat_count += 1
            last_time = now
    return stage, counter, cheat_count, last_time, cheat

def situp_counter(results, stage, counter, last_time, cheat_count, THRESHOLD=0.1, min_range=0.08):
    nose_y = results.pose_landmarks.landmark[mp_pose.PoseLandmark.NOSE].y
    hip_y = results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_HIP].y
    now = time.time()
    cheat = False

    # Cheat: not enough range
    if abs(nose_y - hip_y) < min_range:
        cheat = True

    # Cheat: rep too fast
    if last_time and (now - last_time < 0.5):
        cheat = True

    if nose_y > hip_y + THRESHOLD:
        stage = "down"
    if nose_y < hip_y - THRESHOLD and stage == "down":
        stage = "up"
        if not cheat:
            counter += 1
            last_time = now
        else:
            cheat_count += 1
            last_time = now
    return stage, counter, cheat_count, last_time, cheat

def pushup_counter(results, stage, counter, last_time, cheat_count, THRESHOLD=0.05, min_range=0.05):
    left_shoulder_y = results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_SHOULDER].y
    left_wrist_y = results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_WRIST].y
    now = time.time()
    cheat = False

    # Cheat: not enough range
    if abs(left_shoulder_y - left_wrist_y) < min_range:
        cheat = True

    # Cheat: rep too fast
    if last_time and (now - last_time < 0.5):
        cheat = True

    if left_shoulder_y > left_wrist_y + THRESHOLD:
        stage = "down"
    if left_shoulder_y < left_wrist_y - THRESHOLD and stage == "down":
        stage = "up"
        if not cheat:
            counter += 1
            last_time = now
        else:
            cheat_count += 1
            last_time = now
    return stage, counter, cheat_count, last_time, cheat

def jumping_jack_counter(results, stage, counter, last_time, cheat_count, THRESHOLD=0.15, min_range=0.12):
    left_wrist_y = results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_WRIST].y
    right_wrist_y = results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_WRIST].y
    left_ankle_y = results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_ANKLE].y
    right_ankle_y = results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_ANKLE].y
    avg_wrist_y = (left_wrist_y + right_wrist_y) / 2
    avg_ankle_y = (left_ankle_y + right_ankle_y) / 2
    now = time.time()
    cheat = False

    # Cheat: not enough range
    if abs(avg_wrist_y - avg_ankle_y) < min_range:
        cheat = True

    # Cheat: rep too fast
    if last_time and (now - last_time < 0.5):
        cheat = True

    if avg_wrist_y < avg_ankle_y - THRESHOLD:
        stage = "up"
    if avg_wrist_y > avg_ankle_y + THRESHOLD and stage == "up":
        stage = "down"
        if not cheat:
            counter += 1
            last_time = now
        else:
            cheat_count += 1
            last_time = now
    return stage, counter, cheat_count, last_time, cheat

# --- Exercise Config ---
EXERCISES = {
    "squat": {"name": "Squat", "counter": squat_counter, "color": (0,255,0), "key": "s"},
    "situp": {"name": "Situp", "counter": situp_counter, "color": (255,255,0), "key": "i"},
    "curl": {"name": "Bicep Curl", "counter": bicep_curl_counter, "color": (255,0,0), "key": "c"},
    "pushup": {"name": "Pushup", "counter": pushup_counter, "color": (0,255,255), "key": "p"},
    "jack": {"name": "Jumping Jack", "counter": jumping_jack_counter, "color": (255,0,255), "key": "j"},
}

# --- Main Webcam Loop ---
cap = cv2.VideoCapture(0)
mp_pose = mp.solutions.pose
exercise = "squat"
counter = 0
cheat_count = 0
stage = None
last_time = None

instructions = "Press: S=Squat | I=Situp | C=Curl | P=Pushup | J=Jack | Q=Quit"

with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False
        results = pose.process(image)
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

        mp.solutions.drawing_utils.draw_landmarks(
            image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)

        cheat = False
        if results.pose_landmarks:
            stage, counter, cheat_count, last_time, cheat = EXERCISES[exercise]["counter"](
                results, stage, counter, last_time, cheat_count
            )

        # Display exercise name and counter
        cv2.putText(image, f'{EXERCISES[exercise]["name"]}: {counter}', (10, 60),
                    cv2.FONT_HERSHEY_SIMPLEX, 1, EXERCISES[exercise]["color"], 2)
        # Display cheat count
        cv2.putText(image, f'Cheat Reps: {cheat_count}', (10, 100),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0,0,255), 2)
        # Display instructions
        cv2.putText(image, instructions, (10, 30),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255,255,255), 2)
        # Cheat warning
        if cheat:
            cv2.putText(image, "Cheat Detected!", (10, 140),
                        cv2.FONT_HERSHEY_SIMPLEX, 1, (0,0,255), 3)

        cv2.imshow("Exercise Counter", image)
        key = cv2.waitKey(10) & 0xFF
        if key == ord('q'):
            break
        for ex, cfg in EXERCISES.items():
            if key == ord(cfg["key"]):
                exercise = ex
                counter = 0
                cheat_count = 0
                stage = None
                last_time = None

cap.release()
cv2.destroyAllWindows()
```
